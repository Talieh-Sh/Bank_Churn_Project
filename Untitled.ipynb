```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python

data_file = Path("train.csv")
```

```python
data_file_df = pd.read_csv(data_file)
data_file_df
```

```python
#Checking for the Duplicate values
duplicate_values = data_file_df.duplicated().sum()
duplicate_values
```

```python

#show datatypes
data_file_df.dtypes
```

```python
data_df = data_file_df.dropna()
data_df
```

```python
#removing customers over the age of 60 
df_60 = data_df.loc[data_df['Age'] <= 60.0]
df_60
```

```python
#Removing rows with Surnames that contain a question mark 
df_60 = df_60[~df_60['Surname'].str.endswith(('?'))]
df_60
```
